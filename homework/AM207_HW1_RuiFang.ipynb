## Problem 1: Bayes Theorem

**Coding not required**

### Part A:
Your child has been randomly selected for Type I diabetes screening, using a highly accurate new test that boasts of a false positive rate of 1% and a false negative rate of 0%. The prevalence of of Type I diabetes in children is approximately 0.228%. 

1.Should your child test positive, what is the probability that he/she has Type I diabetes?

2.Should you be concerned enough to ask for further testing or treatment for your child? Suppose an independent test with the same false positive/false negative rate is available.

Later, you read online that Type I diabetes is 6 times more prevalent in prematurely born children. 

3.If this statistic is true, what is the probability that your child, who is prematurely born, has Type I diabetes?

4.Subjectively, given the new information, should you be concerned enough to ask for treatment for your child?

**Justify your decisions using your calculations.**


***

### *Solution*

Let random variable $T$ denote a test result. Possible values of $T$ are $pos$, $neg$. 

Let random variable $D$ denote whether or not having Type I diabetes. Possible values of $D$ are $true$, $false$.

As given by the problem, 

$\text{a false positive rate of 1%} \Longrightarrow P(T=pos \mid D=false) = 0.01,$

$\text{a false negative rate of 0%} \Longrightarrow P(T=neg \mid D=true) = 0,$

$\text{prevalence of Type I diabetes in children is approximately 0.228%} \Longrightarrow P(D=true) = 0.00228.$

1.  Use Bayes Theorem, 
    $$
    P(D\mid T) 
    = \frac{P(T\mid D)P(D)}{P(T)}
    = \frac{P(T\mid D)P(D)}{\sum_{D'}P(T, D')}
    = \frac{P(T\mid D)P(D)}{\sum_{D'}P(T\mid D')P(D')}.
    $$
   
    The probability that the child has Type I diabetes should the child test positive is then 
    $$
    \begin{split}
    P(D=true\mid T=pos) 
    &= \frac{P(T=pos\mid D=true)P(D=true)}{P(T=pos\mid D=true)P(D=true)+P(T=pos\mid D=false)P(D=false)}\\
    &= \frac{(1-0) \times 0.00228}{(1-0) \times 0.00228 + 0.01 \times (1 - 0.00228)}\\
    \approx 18.6\%.
    \end{split}
    $$

    The probability is approximately **18.6%**.  

2.  Similarly, define $T_1$, $T_2$ as two random variables representing respectively the result of the first and second test.  

    From Bayes Theorem, 
    $$
    P(D\mid T_1, T_2)
    = \frac{P(D, T_1, T_2)}{P(T_1, T_2)}
    = \frac{P(D, T_1, T_2)}{\sum_{D'}P(T_1, T_2, D')}
    = \frac{P(T_1, T_2 \mid D)P(D)}{\sum_{D'}P(T_1, T_2 \mid D')P(D')}.
    $$
    
    Because the two tests are independent, $P(T_1, T_2 \mid D)=P(T_1\mid D)P(T_2\mid D)$. Thus, 
    
    $$
    P(D\mid T_1, T_2)
    = \frac{P(T_1\mid D)P(T_2\mid D)P(D)}{\sum_{D'}P(T_1\mid D')P(T_2\mid D')P(D')}.
    $$
    
    Given two independent positive tests, the probability of the child having Type I diabetes is 
    $$
    \begin{split}
    P(D=true\mid T_1=pos, T_2=pos)
    &= \frac{P(T_1=pos\mid D=true)P(T_2=pos\mid D=true)P(D=true)}{P(T_1=pos\mid D=true)P(T_2=pos\mid D=true)P(D=true) + P(T_1=pos\mid D=false)P(T_2=pos\mid D=false)P(D=false)}\\
    &= \frac{(1-0)\times (1-0) \times 0.00228}{(1-0)\times (1-0) \times 0.00228 + 0.01 \times 0.01 \times (1-0.00228)}\\
    &\approx 95.8\%.
    \end{split}
    $$
    
    Therefore, if both tests are positive, the probability of the child have diabetes is about **95.8%**. Compared to the probability of having diabetes given one positive test, this drastically decreases the level of uncertainty. So a second test is worth taking. I would be concerned to take further testing for my child. 

3. Introduce another random variable $Pre$ to indicate if a child is prematurely born. Possible values of $Pre$ are $true$, $false$. 
    The sentence that "Type I diabetes is 6 times more prevalent in prematurely born children" tells us $P(D=true \mid Pre=true) = 6\times0.00228.$
    
    Use Bayes Theorem and the assumption that $T$ is independent of $Pre$ when given $D$, 
    $$
    P(D\mid T, Pre) 
    = \frac{P(T, D, Pre)}{P(T, Pre)}
    = \frac{P(T, D, Pre)}{\sum_{D'}P(T, D', Pre)}
    = \frac{P(T\mid D, Pre)P(D\mid Pre)P(Pre)}{\sum_{D'}P(T\mid D', Pre)P(D'\mid Pre)P(Pre)}
    = \frac{P(T\mid D)P(D\mid Pre)}{\sum_{D'}P(T\mid D')P(D'\mid Pre)}.
    $$
    
    Then, given that the child is prematurely born and is tested positive, the probability of the child having Type I diabetes is 
    $$
    \begin{split}
    P(D=true\mid T=pos, Pre=true) 
    &= \frac{P(T=pos\mid D=true)P(D=true\mid Pre=true)}{P(T=pos\mid D=true)P(D=true\mid Pre=true)+P(T=pos\mid D=false)P(D=false\mid Pre=true)}\\
    &= \frac{(1-0) \times 6\times 0.00228}{(1-0) \times 6 \times 0.00228 + 0.01 \times (1 - 6 \times 0.00228)}\\
    &\approx 58.1\%.
    \end{split}
    $$
    
    The probability is approximately **58.1%**.
    
4.  Since 58.1% is relatively high, I would be concerned enough to ask for treatment for my child.

### Part B:

During shopping week, you're trying to decide between two classes based on the criteria that the class must have a lenient grading system. 

Both classes have hard TFs who give assignments lower grades than the work merits. You know from rumors that in one class has 35% of the TF staff are harsh graders and in the other class 15% are harsh graders, but you don't know which class is which.

So, you decide to conduct an experiment: submit an assignment to be graded. Fortunately, both classes offer an optional Homework 0 that is graded as extra credit. Unfortunately, you only have time to complete the problem set for just one of these classes. 

Suppose you randomly pick the Homework 0 from Class A to complete and suppose that you received a grade that you believe is lower than the quality of your work warrents. 

1. Based on this evidence, what is the probability that Class A has the harsher grading system? Which class should you drop based on the results of your experiment (or do you not have sufficient evidence to decide)?


**Justify your decisions using your calculations.**

***

### *Solution*

1.  Let random variable $LG$ denote whether an assignment is given a lower grade than work merits. Possible values of $LG$ are $true$, $false$. 
    
    Let random variable $C$ denote the class. $C$ can be either $Class 1$ or $Class 2$. 
    
    We know that in $Class 1$ 35% of the TF staff are harsh graders, and in $Class 2$ 15% are harsh graders. In other words, $$P(LG=true \mid C=Class 1)=0.35,$$ $$P(LG=true \mid C=Class 2)=0.15.$$ 

    Now we randomly pick a class and submit an assignment. The probability of the class being either class given the assignment grade can be calculated as below, using Bayes Theorem: 
    $$P(C \mid LG) = \frac{P(LG \mid C)P(C)}{P(LG)} = \frac{P(LG \mid C)P(C)}{\sum_{C'}P(LG,C')} = \frac{P(LG \mid C)P(C)}{\sum_{C'}P(LG \mid C')P(C')}.$$

    Suppose the probability of each class being chosen is 50% ($P(C=Class 1)=P(C=Class 2)=0.5$). Then if we received a lower grade from Class A, the probability that Class A has the harsher grading system ($C=Class1$) is
    $$
    \begin{split}
    P(C=Class 1 \mid LG=true) 
    &= \frac{P(LG=true \mid C=Class 1)P(C=Class 1)}{P(LG=true \mid C=Class 1)P(C=Class 1)+P(LG=true \mid C=Class 2)P(C=Class 2)}\\
    &= \frac{0.35 \times 0.5}{0.35 \times 0.5 + 0.15 \times 0.5}\\
    &= 70\%.
    \end{split}
    $$
    
    Therefore, based on the evidence, the probability that Class A has the harsher grading system is **70%**. I would drop Class A and take the other class.

## Problem 2: Maximum Likelihood Estimates

**Coding required for Part B only**

Suppose you observe the following data set $\mathbf{x}^{(0)} = (0.5, 2.5), \mathbf{x}^{(1)} = (3.2, 1.3), \mathbf{x}^{(2)} = (2.72, 5.84), \mathbf{x}^{(3)}= (10.047, 0.354)$. By convention, for any vector $\mathbf{x}$, we will denote the first component of $\mathbf{x}$ by $x_{1}$ and the second component by $x_{2}$. Suppose that the data is drawn from the same two-dimensional probability distribution with pdf $f_X$, that is, $\mathbf{x}^{(i)} \overset{iid}{\sim} f_X$, where
$$
f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\}.
$$
You should assume that $\lambda_1, \lambda_0 > 0$ and that $f_X$ is supported on the nonnegative quandrant of $\mathbb{R}^2$ (i.e. $f_X$ is zero when either component is negative). 

### Part A:

What are the values for $\lambda_0$ and $\lambda_1$ that maximizes the likelihood of the observed data? 

**Support your answer with full and rigorous analytic derivations.**

***

### *Solution*

First, the pdf must be normalized. Setting the integral of the pdf over the nonnegative quadrant of $\mathbb{R}^2$ to be equal to 1: 

$$
\begin{split}
\int_0^\infty \int_0^\infty f_X\left(x_1, x_2\right) dx_1 dx_2 
&= \int_0^\infty \int_0^\infty 4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 \left(x^2_{1} + x^2_{2}\right) \right\} dx_1 dx_2 \\
&= 4\lambda_1^2 \int_0^\infty x_{1} \mathrm{exp} \left\{-\lambda_0 x^2_{1} \right\} dx_1 \int_0^\infty x_{2} \mathrm{exp} \left\{-\lambda_0 x^2_{2} \right\} dx_2 \\
&= 4\lambda_1^2 \left. \left[ -\frac{1}{2\lambda_0} \mathrm{exp} \left\{-\lambda_0 x^2_{1} \right\} \right] \right|_0^\infty \left. \left[ -\frac{1}{2\lambda_0} \mathrm{exp} \left\{-\lambda_0 x^2_{2} \right\} \right] \right|_0^\infty \\
&= \frac{\lambda_1^2 }{\lambda_0^2} \\
&= 1.
\end{split}
$$

Hence, we find a relationship between $\lambda_1$ and $\lambda_0$. Since we assume $\lambda_1, \lambda_0 > 0$, we must have $\lambda_1 = \lambda_0$. 

Let $\lambda_1 = \lambda_0 = \lambda$. The likelihood of the observed data is given by  
$$\mathcal{L} = \prod_{i=0}^3 f_X (\mathbf{x}^{(i)}) = \prod_{i=0}^3 4\lambda^2 x^{(i)}_{1}x^{(i)}_{2} \mathrm{exp} \left\{-\lambda \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right) \right\}.$$

And the log likelihood is
$$
\log \mathcal{L} = \sum_{i=0}^3 \left[ \log \left( 4\lambda^2 \right) + \log \left( x^{(i)}_{1}x^{(i)}_{2} \right) - \lambda \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right) \right] = 8 \log(4\lambda) + \sum_{i=0}^3 \log \left( x^{(i)}_{1}x^{(i)}_{2} \right) - \sum_{i=0}^3 \lambda \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right). 
$$

To find the optimal $\lambda$ that maximizes the likelihood, take derivative of the log likelihood with respect to $\lambda$ and set it to 0: 
$$
\frac{\partial}{\partial \lambda} \log \mathcal{L} = \frac{8}{\lambda} - \sum_{i=0}^3 \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right) = 0.
$$

Solving for $\lambda$ gives 
$$
\lambda = \frac{8}{\sum_{i=0}^3 \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right)}.
$$

Therefore, $\lambda_1 = \lambda_0 = \frac{8}{\sum_{i=0}^3 \left( (x^{(i)}_{1})^2 + (x^{(i)}_{2})^2\right)} \approx 0.0497$ maximizes the likelihood of the observed data $\mathbf{x}^{(0)} = (0.5, 2.5)$, $\mathbf{x}^{(1)} = (3.2, 1.3)$, $\mathbf{x}^{(2)} = (2.72, 5.84)$, $\mathbf{x}^{(3)}= (10.047, 0.354)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

# Set plotting style 
sns.set(palette='deep', color_codes=True)

# Calculate lambda from given data 
data_p2 = np.array([[0.5, 2.5], [3.2, 1.3], [2.72, 5.84], [10.047, 0.354]])
lmd = 8 / np.sum(data_p2**2)

# Define distribution function 
def f_x(x1, x2, lmd):
    return 4 * lmd**2 * x1 * x2 * np.exp(-lmd * (x1**2 + x2**2))

# Prepare to plot distribution function 
x1 = np.linspace(0, 11, 200)
x2 = np.linspace(0, 11, 200)
X1, X2 = np.meshgrid(x1, x2)

# Make contour plot 
plt.figure()
plt.contourf(X1, X2, f_x(X1, X2, lmd), 15, cmap='viridis')
plt.colorbar()
plt.scatter(data_p2[:,0], data_p2[:,1], color='w', marker='x', s=20, label='data')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Contour plot of probability distribution and data points')
plt.axes().set_aspect('equal')
plt.legend(loc='best', frameon=True)
plt.show()

# Make surface plot
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X1, X2, f_x(X1, X2, lmd), cmap='viridis', linewidth=0.2)
fig.colorbar(surf, shrink=1.0)
ax.view_init(30, 250)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title('Surface plot of probability distribution')
plt.show()
```

### Part (B): Estimation using bootstrap

1. Empirically estimate the mean of the distribution $f_{x_1}$. Estimate, also the SE (standard error) of the estimate.

1. Empirically estimate the standard deviation of the distribution $f_{x_2 \mid x_1}$, for $x_1 \in [3.99, 4.01]$. Estimate, also the SE (standard error) of the estimate.

1. Given the SE, How many digits in your standard deviation estimate are significant? Explain why.

In obtaining estimates for this problem we want you to 

- define a function called *get_bootstrap_sample(dataset)* to generate each bootstrap sample
- and then another function *perform_bootstrap(dataset)* to generate all the samples. 

They should both take as parameters the dataset from which you'll be drawing samples. *perform_bootstrap* should call *get_bootstrap_sample* and return a sequence of bootstrap samples.  *get_bootstrap_sample* should return an individual bootstrap sample.

    

*** 

### *Solution* 

```python
# Function to generate each bootstrap sample
def get_bootstrap_sample(dataset, sample_size):
    return np.random.choice(dataset, sample_size, replace=True)

# Function to generate all the samples 
def perform_bootstrap(dataset, sample_size, num_bootstraps):
    
    # Initialize array to store all the samples 
    bootstrap_samples = np.zeros((num_bootstraps, sample_size))
    
    # Repeat calling get_bootstrap_sample to generate samples for num_bootstraps times 
    for i in range(num_bootstraps):
        bootstrap_samples[i] = get_bootstrap_sample(dataset, sample_size) 
    
    return bootstrap_samples
```

```python
# 1) Estimate the mean of the distribution f_x1 
x1_bootstrap_samples = perform_bootstrap(x1, x1.shape[0], 10000)
x1_mean_samples = np.mean(x1_bootstrap_samples, axis=1)

print('Mean estimate of f_x1:\n  ', np.mean(x1_mean_samples))
print('SE of estimate:\n  ', np.std(x1_mean_samples))

# Plot histogram 
plt.figure()
sns.distplot(x1_mean_samples, kde=False, ax=plt.axes())
plt.xlabel('Bootstraped mean of $f_{x_1}$')
plt.ylabel('Count')
plt.title('Histogram of bootstraped means')
plt.show()
```

**By performing 10000 bootstraps, we obtain that the empirical estimate of the mean of the distribution $f_{x_1}$ is $3.993 \pm 0.007$**.

```python
# 2) Estimate the standard deviation of the distribution f_x2|x1 for x1 in [3.99, 4.01] 
x2_given_x1_bootstrap_samples = perform_bootstrap(x2_given_x1, x2_given_x1.shape[0], 10000)
x2_given_x1_std_samples = np.std(x2_given_x1_bootstrap_samples, axis=1)

print('Standard deviation estimate of f_x2|x1 for x1 in [3.99, 4.01]:\n  ', np.mean(x2_given_x1_std_samples))
print('SE of estimate:\n  ', np.std(x2_given_x1_std_samples))

# Plot histogram 
plt.figure()
sns.distplot(x2_given_x1_std_samples, kde=False, ax=plt.axes())
plt.xlabel('Bootstraped standard deviation of $f_{x_2 \mid x_1}$ for $x_1 \in [3.99, 4.01]$')
plt.ylabel('Count')
plt.title('Histogram of bootstraped standard deviations')
plt.show()
```

**By performing 10000 bootstraps, we obtain that the empirical estimate of the standard deviation of the distribution $f_{x_2 \mid x_1}$, for $x_1 \in [3.99, 4.01]$ is $0.73 \pm 0.04$**. 

Given the SE, we can be very confident that the standard deviation lies between 0.69 and 0.77 most of the time. However, we can not be this certain about any digit after the second digit. Thus, two digits in the standard deviation estimate are significant.

## Problem 4: Missing Data

**Coding required**

Missing data is a very important topic in statistics and machine learning. Let us begin to explore how to handle missing data in our analysis.  We'll be working with a dataset from the UCI Machine Learning Repository that uses a variety of wine chemical predictors to classify wines grown in the same region in Italy.  Each line represents 13 (mostly chemical) predictors of the response variable wine class, including things like alcohol content, hue, and phenols.  Unfortunately some of the predictor values were lost in measurement. Please load `wine_quality_missing.csv`.  (If pandas makes your life easier, feel free to use it).

1.  One way to handle missing data is to just totally ignore it by removing any rows that have any missing values.  This is called *drop imputation*.  Use drop imputation on our wine quality dataset.  How many rows does our imputed dataset have?

1. Another way to handle missing data is to replace any missing value with the mean of the non-missing values in that **column**.  This is called *mean imputation*.  How many rows does our mean imputed dataset have?

1.  Empirically estimate the pdf of the distribution of the sample mean of the ash predictor under drop imputation and mean imputation. Use the perform_bootstrap function from problem 3.

1. Compare the standard errors (SE) of the distributions under drop imputation and mean imputation in the previous question.  Can you suggest one or two reasons why they may be different?

***

### *Solution*

```python
# Load in dataset 
df = pd.read_csv('wine_quality_missing.csv')

# Handle missing data with drop imputation 
drop_imputed_df = df.dropna(axis=0, how='any')

# Handle missing data with mean imputation
mean_imputed_df = df.fillna(df.mean())

# Print number of rows in each imputed dataset 
print('Drop imputed dataset has %s rows.' % drop_imputed_df.shape[0])
print('Mean imputed dataset has %s rows.' % mean_imputed_df.shape[0])
```

```python
# Estimate sample mean of the ash predictor under drop imputation and mean imputation using bootstrap
bootstrap_samples_drop_imputed = perform_bootstrap(drop_imputed_df['Ash'], drop_imputed_df.shape[0], 10000)
bootstrap_samples_mean_imputed = perform_bootstrap(mean_imputed_df['Ash'], mean_imputed_df.shape[0], 10000)
drop_imputed_means = np.mean(bootstrap_samples_drop_imputed, axis=1)
mean_imputed_means = np.mean(bootstrap_samples_mean_imputed, axis=1)

# Plot distributions using kernel density estimate with Gaussian kernels 
plt.figure(figsize=(8,6))
sns.distplot(drop_imputed_means, ax=plt.axes(), label='under drop imputation')
sns.distplot(mean_imputed_means, ax=plt.axes(), label='under mean imputation')
plt.xlabel('Bootstraped mean of the ash predictor')
plt.ylabel('Density')
plt.title('Distributions of the sample mean of the ash predictor under different imputations')
plt.legend(loc='best')
plt.show()

# Print mean estimates and standard errors (SE) 
print('Under drop imputation:\n  mean = %s, SE = %s' % (np.mean(drop_imputed_means), np.std(drop_imputed_means)))
print('Under mean imputation:\n  mean = %s, SE = %s' % (np.mean(mean_imputed_means), np.std(mean_imputed_means)))
```

The sample mean estimate of the ash predictor is $2.37 \pm 0.05$ under drop imputation, and is $2.37 \pm 0.02$ under mean imputation. 

The SE of the sample mean estimate under drop imputation is bigger than that under mean imputation. This is because the drop imputed dataset is smaller than the mean imputed dataset. If the dataset is small, the sample size used in bootstrap, which is equal to the dataset size, is also small. The mean calculated on a smaller dataset has greater variance than the mean calculated on a larger dataset. Therefore, the standard error (SE) of the sample mean estimate is bigger under drop imputation.
